## **Import LangChain Model and Initialize OpenRouter Model**

```python
from langchain_openai import ChatOpenAI
import os
from dotenv import load_dotenv

load_dotenv()

llm = ChatOpenAI(
    model="openai/gpt-4o-mini",
    api_key=os.getenv("OPENROUTER_API_KEY"),
    base_url="https://openrouter.ai/api/v1"
)
```

```python
response = llm.invoke("Hello, introduce yourself in one sentence.")

print(response.content)
```

## **Test Structured Reservation Prompt**


```python
prompt = """
Extract reservation details from this message:

'I need a table for 5 people tomorrow at 7 PM under the name Shoaib'

Return:
- Name
- Number of guests
- Date
- Time
"""

response = llm.invoke(prompt)

print(response.content)
```
