## **Imports**

```python
import json
from datetime import datetime
```

## **Example Reservation Data**


```python
reservation_data = {
    "name": "Shoaib",
    "date": "Tomorrow",
    "time": "8 PM",
    "guests": 4
}

print(reservation_data)
```

## **Create Admin Request Generator**

```python
def generate_admin_request(reservation):
    
    message = f"""
    NEW RESERVATION REQUEST
    
    Customer Name : {reservation['name']}
    Reservation Date : {reservation['date']}
    Reservation Time : {reservation['time']}
    Number of Guests : {reservation['guests']}
    
    Please approve or reject this reservation.
    """
    
    return message
```

## **Generate Approval Message**

```python
admin_message = generate_admin_request(reservation_data)

print(admin_message)
```

## **Simulate Human Admin Response**

```python
admin_response = input("Enter admin decision (approve/reject): ")

print("Admin Decision:", admin_response)
```

## **Process Admin Decision**

```python
def process_admin_decision(decision):
    
    decision = decision.lower()
    
    if decision == "approve":
        return {
            "status": "approved",
            "message": "Reservation approved by administrator."
        }
    
    elif decision == "reject":
        return {
            "status": "rejected",
            "message": "Reservation rejected by administrator."
        }
    
    else:
        return {
            "status": "invalid",
            "message": "Invalid admin response."
        }
```

## **Get Final Result**

```python
final_result = process_admin_decision(admin_response)

print(final_result)
```

## **Add Timestamp (Professional Touch)**

```python
final_result["processed_at"] = str(datetime.now())

print(final_result)
```

## **Simulate Full Workflow**

```python
print("Customer Reservation Submitted")
print()

print(admin_message)
print()

print("Admin Decision:", admin_response)
print()

print("Final Result:")
print(final_result)
```
