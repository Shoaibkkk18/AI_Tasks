## **Imports**

```python
import json
from datetime import datetime
```

## **Sample Reservation**

```python
reservation_data = {
    "name": "Shoaib",
    "date": "Tomorrow",
    "time": "8 PM",
    "guests": 4
}

print(reservation_data)
```

## **Reservation Validation Function**

```python
def validate_reservation(reservation):

    required_fields = [
        "name",
        "date",
        "time",
        "guests"
    ]

    # Check missing fields
    for field in required_fields:

        if field not in reservation:
            return False, f"Missing field: {field}"

    # Validate name
    if not reservation["name"].strip():
        return False, "Customer name cannot be empty"

    # Validate guests
    if reservation["guests"] <= 0:
        return False, "Guests must be greater than 0"

    return True, "Reservation is valid"
```

## **Test Valid Reservation**

```python
is_valid, message = validate_reservation(
    reservation_data
)

print(is_valid)
print(message)
```

## **Test Negative Guest Count**

```python
bad_reservation = {
    "name": "Shoaib",
    "date": "Tomorrow",
    "time": "8 PM",
    "guests": -2
}

is_valid, message = validate_reservation(
    bad_reservation
)

print(is_valid)
print(message)
```

## **Test Empty Name**

```python
bad_reservation = {
    "name": "",
    "date": "Tomorrow",
    "time": "8 PM",
    "guests": 4
}

is_valid, message = validate_reservation(
    bad_reservation
)

print(is_valid)
print(message)
```

## **Test Missing Field**

```python
bad_reservation = {
    "name": "Shoaib",
    "time": "8 PM",
    "guests": 4
}

is_valid, message = validate_reservation(
    bad_reservation
)

print(is_valid)
print(message)
```

## **Admin Response Validation**

```python
def validate_admin_response(response):

    valid_responses = [
        "approve",
        "reject"
    ]

    response = response.strip().lower()

    if response not in valid_responses:
        return False, "Invalid admin response"

    return True, "Valid admin response"
```

## **Test Invalid Admin Response**

```python
is_valid, message = validate_admin_response(
    "maybe"
)

print(is_valid)
print(message)
```

## **Test Valid Admin Response**

```python
is_valid, message = validate_admin_response(
    "approve"
)

print(is_valid)
print(message)
```

## **Safe JSON Loader**

```python
def safe_json_load(json_text):

    try:

        data = json.loads(json_text)

        return True, data

    except json.JSONDecodeError as e:

        return False, str(e)
```

## **Test Invalid JSON**

```python
bad_json = "{ name: Shoaib }"

success, result = safe_json_load(
    bad_json
)

print(success)
print(result)
```

## **Test Valid JSON**

```python
good_json = '''
{
    "name": "Shoaib"
}
'''

success, result = safe_json_load(
    good_json
)

print(success)
print(result)
```
