## **Imports**

```python
from flask import Flask, jsonify, request
from datetime import datetime
```

## **Create Flask App**

```python
app = Flask(__name__)

print("Flask app initialized")
```

## **Create Global Reservation Store**

```python
reservation_data = {
    "name": "Shoaib",
    "date": "Tomorrow",
    "time": "8 PM",
    "guests": 4
}

reservation_status = {
    "status": "pending",
    "processed_at": None
}

print("Reservation system ready")
```

## **Reservation Endpoint**

```python
@app.route("/reservation", methods=["GET"])
def get_reservation():
    
    return jsonify({
        "reservation": reservation_data,
        "status": reservation_status
    })
```

## **Approve Endpoint**

```python
@app.route("/approve", methods=["POST"])
def approve_reservation():
    
    reservation_status["status"] = "approved"
    reservation_status["processed_at"] = str(datetime.now())
    
    return jsonify({
        "message": "Reservation approved",
        "status": reservation_status
    })
```

## **Reject Endpoint**

```python
@app.route("/reject", methods=["POST"])
def reject_reservation():
    
    reservation_status["status"] = "rejected"
    reservation_status["processed_at"] = str(datetime.now())
    
    return jsonify({
        "message": "Reservation rejected",
        "status": reservation_status
    })
```

## **Status Endpoint**

```python
@app.route("/status", methods=["GET"])
def get_status():
    
    return jsonify(reservation_status)
```

```python
@app.route("/")
def home():
    return "Human-in-the-Loop Reservation API Running"
```

## **Run Flask App**

```python
from threading import Thread

def run_flask():
    app.run(
        host="127.0.0.1",
        port=5000,
        debug=False,
        use_reloader=False
    )

flask_thread = Thread(target=run_flask)
flask_thread.start()

print("Flask server running in background")
```
