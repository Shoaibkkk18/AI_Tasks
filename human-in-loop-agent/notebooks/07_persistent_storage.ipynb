## **Imports**

```python
import json
import os
from datetime import datetime
```

## **Create Data Folder**

```python


os.makedirs("../data", exist_ok=True)

print("Root data folder created")
```

## **Create Database File Path**

```python
DATABASE_FILE = "../data/reservations.json"

print(DATABASE_FILE)
```

## **Initialize JSON Database**

```python
if not os.path.exists(DATABASE_FILE):
    
    with open(DATABASE_FILE, "w") as file:
        json.dump([], file)

print("Database initialized")
```

## **Create Sample Reservation**


```python
reservation_data = {
    "name": "Shoaib",
    "date": "Tomorrow",
    "time": "8 PM",
    "guests": 4,
    "status": "approved",
    "processed_at": str(datetime.now())
}

print(reservation_data)
```

## **Create Save Function**


```python
def save_reservation(reservation):

    print("FUNCTION STARTED")

    # Read existing reservations
    with open(DATABASE_FILE, "r") as file:
        reservations = json.load(file)

    print("CURRENT DATA:", reservations)

    # Add new reservation
    reservations.append(reservation)

    print("UPDATED DATA:", reservations)

    # Write updated reservations
    with open(DATABASE_FILE, "w") as file:
        json.dump(reservations, file, indent=4)

    print("Reservation saved successfully")
```

## **Save Reservation**

```python
save_reservation(reservation_data)
```

## **Read Reservations**

```python
with open(DATABASE_FILE, "r") as file:
    all_reservations = json.load(file)

print(all_reservations)
```

## **Pretty Print Reservations**

```python
for reservation in all_reservations:
    
    print("==========")
    print("Customer:", reservation["name"])
    print("Date:", reservation["date"])
    print("Time:", reservation["time"])
    print("Guests:", reservation["guests"])
    print("Status:", reservation["status"])
    print("Processed At:", reservation["processed_at"])
```

```python
with open(DATABASE_FILE, "r") as file:
    content = file.read()

print(content)
```
